```python
######Creación del un código para ordenar las simulaciones según los parámetros de las trampas utilizados
import os
import re
import shutil

path='E:\\1_TFG\\TFG\\Simulaciones\\Obstaculos\\'
new_path_base='E:\\1_TFG\\TFG\\Simulaciones\\Tartarus-Punto\\'

folder1=os.listdir(path)
for folder in folder1:
    folder1_path=path+folder+'\\'
    #gamma: Indica el ancho del anillo
    gamma=re.split('_', folder)[-1] 
    #delta: Indica la diferencia entre la longitud de los lados del obstaculo en forma de V
    delta=re.split('[-_]', folder)[22]
    """aa y bb: Indican el  ́angulo de cada una de las ramas de los chevrons con respecto a la 
       radial de las circunferencias en las que se sit ́uan los mismo"""
    aa=re.split('[-_]', folder)[16]
    bb=re.split('[-_]', folder)[18]
    #L1: Indica la longitud de una de las ramas del obstaculo en forma de V
    L1=re.split('[-_]', folder)[24]
    #Nf: Indica el numero de chevrons que habra en cada una de las circunferencias
    Nf=re.split('[-_]',folder)[14]
    #R1: Indica la distancia del centro de la caja a la circunferencia exterior de chevrons
    R1=re.split('[-_]', folder)[20]
    #vp: Velocidad de propulsión
    v=re.split('[-_]', folder)[5]
    #lp: Longitud de persistencia
    lp=re.split('[-_]', folder)[7]
    #dt: Paso de tiempo
    dt=re.split('[-_]', folder)[12]
    
    alpha=float(vp)*float(dt)/float(lp)
    conf_run_files=os.listdir(folder1_path)
    for conf_run in conf_run_files:
        R1_path=new_path_base+R1+'\\'
        if not os.path.exists(R1_path):
            os.mkdir(R1_path)
        gamma_path=R1_path+gamma+'\\'
        if not os.path.exists(gamma_path):
            os.mkdir(gamma_path)
        aa_path=gamma_path+aa+'\\'
        if not os.path.exists(aa_path):
            os.mkdir(aa_path)  
        bb_path=aa_path+bb+'\\'
        if not os.path.exists(bb_path):
            os.mkdir(bb_path)
        delta_path=bb_path+delta+'\\'
        if not os.path.exists(delta_path):
            os.mkdir(delta_path)
        L1_path=delta_path+L1+'\\'
        if not os.path.exists(L1_path):
            os.mkdir(L1_path)
        Nf_path=L1_path+Nf+'\\'
        if not os.path.exists(Nf_path):
            os.mkdir(Nf_path)
        lp_path=Nf_path+lp+'\\'
        if not os.path.exists(lp_path):
            os.mkdir(lp_path)
        v_path=lp_path+vp+'\\'
        if not os.path.exists(v_path):
            os.mkdir(v_path)
        shutil.copy(folder1_path+conf_run, v_path+conf_run)
```
